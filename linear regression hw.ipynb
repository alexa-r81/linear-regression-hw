```python
import pandas as pd
file = pd.read_csv('Diamonds Prices2022.csv')
file.head()
```

```python
file.describe()
```

```python
carat = file['carat']
price = file['price']

data = file[['carat', 'price']]

import matplotlib.pyplot as plt

plt.figure(figsize = (20,20))
plt.ylabel('carat')
plt.xlabel('price')
plt.title('price vs carat')
plt.scatter(price, carat, color = 'pink')
```

```python
import numpy as np
import statsmodels.api as sm
```

```python
x_ary = file['price']
x_with_cnst = sm.add_constant(x_ary)
model = sm.OLS(file['carat'], x_with_cnst)
results = model.fit()
b,m = results.params;
print(m)
print(b)
```

```python
print(results.summary())
```

```python
plt.figure(figsize = (20,20))
plt.scatter(file['price'], file['carat'], color = 'pink')
plt.plot(x_ary, m*x_ary+b, color = 'black', linewidth = 5)
```

```python
import scipy.stats

slope, intercept, r, p, se = scipy.stats.linregress(price, carat)

print(slope, intercept)
```

```python
plt.figure(figsize = (20,20))
plt.scatter(file['price'], file['carat'], color = 'pink')
plt.plot(x_ary, slope*x_ary+intercept, color = 'black', linewidth = 5)
```
